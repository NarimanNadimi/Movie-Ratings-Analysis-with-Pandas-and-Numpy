<h2> اهداف:

<br><h3>2. محاسبه تعداد کل فیلم‌ها و کاربران

```python
import pandas as pd

# خواندن داده
df = pd.read_csv('movies.csv')

# اضافه کردن رکورد بیشتر برای The Matrix
new_rows = pd.DataFrame([
    {'MovieID':1, 'Title':'The Matrix', 'Genre':'Sci-Fi', 'UserID':106, 'Rating':5},
    {'MovieID':1, 'Title':'The Matrix', 'Genre':'Sci-Fi', 'UserID':107, 'Rating':4}
])

# ترکیب با دیتافریم اصلی
df = pd.concat([df, new_rows], ignore_index=True)
df.head()
```

<div dir=rtl align=center>
<h2>بررسی داده‌ها
<br><h3>- ستون‌ها: MovieID, Title, Genre, UserID, Rating
<br><h3>- هدف: تحلیل امتیازات کاربران و فیلم‌ها


```python
# تعداد کل فیلم‌ها
num_movies = df['MovieID'].nunique()
print(f"Total Movies: {num_movies}")

#also :
#num_movies=df['MovieID'].count() --> output:np.int64(20)

# تعداد کاربران
num_users = df['UserID'].nunique()
print(f"Total Users: {num_users}")

# میانگین امتیاز هر فیلم
avg_rating = df.groupby('Title')['Rating'].mean().sort_values(ascending=False)
print("\nTop Rated Movies:")
print(avg_rating.head())
```

<div dir=rtl align=center>
<h2>
تمرین:
<h3>
<br>1. بیشترین فیلمی که بیشترین امتیاز داده شده را دارد پیدا کنید.
<br>2. تعداد فیلم‌ها در هر ژانر را محاسبه کنید.
<br>3. کاربران با بیشترین میانگین امتیاز را پیدا کنید.


```python
# بیشترین فیلم امتیاز داده شده
top_rated_count = df.groupby('Title')['Rating'].count().sort_values(ascending=False)
print("\nMovies with most ratings:")
print(top_rated_count.head())

# تعداد فیلم‌ها در هر ژانر
genre_count = df['Genre'].value_counts()
print("\nNumber of movies per genre:")
print(genre_count)

# کاربران با بیشترین میانگین امتیاز
top_users = df.groupby('UserID')['Rating'].mean().sort_values(ascending=False)
print("\nTop users by average rating:")
print(top_users.head())
```
